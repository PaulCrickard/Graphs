### Definition 5:
A graph is an object consisting of two sets called its vertex set and its edge set. The vertex set is a finite nonempty set.
The edge set may be empty, but otherwise its elements are two-element subsets of the vertex set.

```python
G = nx.Graph()
v=set(['A','B','C'])
e=set([('A','B'),('B','C'),('A','C')])
G.add_nodes_from(v)
G.add_edges_from(e)
nx.draw(G,with_labels = True)
```

```python
# Empty Edges
Gempty = nx.Graph()
v=set(['A','B','C'])
e=set([])
Gempty.add_nodes_from(v)
Gempty.add_edges_from(e)
nx.draw(Gempty,with_labels = True)
```

### Definition 6: 
The elements of the vertex set of a graph are called vertices and the lements of the edge set are called edges. We shall denote the number of vertices by v and the number of edges by e.

```python
G = nx.Graph()
v=set(['A','B','C','D'])
e=set([('A','B'),('B','C'),('A','C')])
G.add_nodes_from(v)
G.add_edges_from(e)

print("n = "+str(G.number_of_nodes()))
print("e = "+str(G.number_of_edges()))
```

### Definition 7:
If (X,Y) is an edge of a graph we say (X,Y) joins or connects the vertices X an Y, and that X and Y are adjacent to one another. The edge (X,Y), is incident to each X and Y, and each of X and Y is incident to (X,Y). Two edges incident to the same vertex are called adjacent edges. A vertex is incident to no edges at all is isolated.

```python
G7 = nx.Graph()
v=set(['P','Q','R','S','T','U'])
e=set([('P','Q'),('P','R'),('Q','R'),('S','U')])
G7.add_nodes_from(v)
G7.add_edges_from(e)
nx.draw(G7,with_labels = True)
```

```python
print([n for n in G7.neighbors('P')])
print([n for n in G7.neighbors('U')])
print(list(nx.isolates(G7)))
```

### Definition 8:
We will say that two graphs are equal if they have equal vertex sets and equsl edge sets. And we will say that two graph diagrams are equal if they represent equal vertex sets and equal edge sets.

```python
G8 = nx.Graph()
v=set(['P','Q','R','S','T','U'])
e=set([('P','Q'),('P','R'),('Q','R'),('S','U')])
G8.add_nodes_from(v)
G8.add_edges_from(e)

G8b = nx.Graph()
v=set(['T','P','R','Q','U','S'])
e=set([('Q','R'),('R','P'),('U','S'),('P','Q')])
G8b.add_nodes_from(v)
G8b.add_edges_from(e)

nx.is_isomorphic(G8,G8b)
```

### Definition 10:
If v is a positive integer, the null graph on v vertices, denoted "Nv", is the graph having vertex set (1,2,3,...,v) and no edges

```python
Nv = nx.empty_graph(3,create_using=nx.DiGraph())
nx.draw(Nv,with_labels = True)
```

### Definition 11:

### Definition 14:

    * to each element of B there has been associated a single element of A.

### Definition 16:
Two grapsh are said to be isomorphic if there exists between their vertex sets a one-to-one orrespondence having the property that whenever two vertices are adjacent in either graph, the corresponding two vertices are adjacent in the other graph. Scuh a one-to-one correspondence is called an isomorphism.

```python
G = nx.Graph()
v=set(['P','Q','R','S','T','U'])
e=set([('P','Q'),('P','R'),('Q','R'),('S','U')])
G.add_nodes_from(v)
G.add_edges_from(e)

H = nx.Graph()
v=set([1,2,3,4,5,6])
e=set([(1,2),(1,3),(2,3),(4,6)])
H.add_nodes_from(v)
H.add_edges_from(e)

nx.is_isomorphic(G,H)
```
